```python
ax = sns.countplot(x='Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount', hue='Gender', data=Sales_gen, palette="tab10", dodge=False, legend=False)
```

```python
ax = sns.countplot(data = df, x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age group
sales_gen = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', hue='Age Group', data=sales_gen, palette="tab10", legend=False)
```

  from above graphs we can see that most of the buyers are of age group between 26-35 yrs female

```python
sales_gen = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', hue='State', data=sales_gen, palette='tab10', dodge=False, legend=False)
```

```python
# total amount/sales from top  states
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', hue='State', data=sales_state, palette='tab10', dodge=False, legend=False)
```

From the above graphs we can see that unexpectedly most of the orders from Uttar pradesh , Maharastra,Karnataka respectively.

#Marotal Status

```python
ax = sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize': (4, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the size of the plot
sns.set(rc={'figure.figsize': (6, 5)})

# Create the bar plot
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=sales_state)
```

*From the above table you can see most of the buyer's are married (woman) and have high purchesing power*

#Occupation

```python
sns.set(rc={'figure.figsize': (20, 5)})

# Choose a color palette, for example, 'tab10'
#color_palette = sns.color_palette('tab10')

ax = sns.countplot(data=df, x='Occupation',hue='Occupation', palette ='tab10', dodge=False, legend=False)

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the size of the plot
sns.set(rc={'figure.figsize': (20, 5)})

# Create the bar plot
sns.barplot(data=sales_state ,x='Occupation', y='Amount',hue='Occupation', palette ='tab10', dodge=False, legend=False)
```

#Product Category

```python
sns.set(rc={'figure.figsize': (20, 5)})

ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette='tab10', dodge=False, legend=False)

for bar in ax.containers:
    ax.bar_label(bar)
```

*Form the above bar we can see most of sold product are Clothing , Food , Electronic&Gadge*

```python
Sales_State  = df.groupby(['Product_Category'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize': (23, 5)})
sns.barplot(data=Sales_State,x = 'Product_Category',y='Amount',hue='Product_Category', palette='tab10', dodge=False, legend=False)
```

*From the above graph we can see that most of the sold product are Food,Clothing&Apparel,Electronic&Gadgets*

```python
Sales_State  = df.groupby(['Product_ID'], as_index=False) ['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=Sales_State,x = 'Product_ID',y='Orders',hue='Product_ID', palette='tab10', dodge=False, legend=False)
```
